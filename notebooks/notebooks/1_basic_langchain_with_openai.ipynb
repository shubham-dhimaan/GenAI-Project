## Static Prompts (Using System Prompts)

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o")

prompts = [
    ("system", "You are an advanced R developer"),
    ("user", "Explain how to calculate mean of a dataset")
] # Feeding system and user prompt

res = llm.invoke(prompts)

print(res.content)
```

## Streaming output in real-time


```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model= 'gemini-2.5-flash-lite')

# res = llm.invoke("Tell me about Quantum Computing in 300 words, explain as if you explaining to a 5 year old kid") # usual invoke, all at one output

for chunk in llm.stream("Tell me about Quantum Computing in 300 words, explain as if you explaining to a 5 year old kid"):
    print(chunk.text, end ='|', flush= True)

# This stream gives iterable chunk (objects of text) that are generated in real time and can be iterated and printed through a for loop
```
