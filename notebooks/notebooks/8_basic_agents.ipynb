```python
from dotenv import load_dotenv
load_dotenv()

from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
```

```python
llm = ChatOpenAI(model='gpt-4')
```

```python
@tool
def add_numbers(a:int, b:int):
    """
    This tool will return the sum of two numbers.

    Args:
        a: Number One
        b: Number Two
    
    """
    return a + b

@tool
def multiply_numbers(a:int, b:int):
    """
    This tool will return the product of two numbers.

    Args:
        a: Number One
        b: Number Two
    
    """
    return a * b

@tool
def divide_numbers(a:int, b:int):
    """
    This tool will return the division of two numbers.

    Args:
        a: Number One
        b: Number Two
    
    """
    return a / b
```

```python
from langchain.agents import create_agent
```

```python
agent = create_agent(
    model= llm,
    tools = [add_numbers, multiply_numbers, divide_numbers],
    system_prompt = "You are a proficient Math teacher and whenevr needed use the given tools for calculations"
)
```

```python
response = agent.invoke({"messages": [{"role": "user", "content": "What is the result of 5 + 10 * 9 / 9"}]})
```

```python
for res in response['messages']:                    # better understanding of the returned output, which is a list inside a dict
    print(res)
    print("\n")

```

```python
print(response['messages'][0].content) 

# the first message in the list is what we sent and then it appends the rest of the interaction (ReAct loop) and gives it back to us
```

```python
print(response['messages'][-1].content)         # printing last item from the list of messages
```
