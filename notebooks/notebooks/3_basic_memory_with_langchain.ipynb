```python
from dotenv import load_dotenv
load_dotenv()

from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model= 'gpt-4o')
```

```python
prompts = [
    {"role": "user", "content": "Hi, my name is Walter White"},
    {"role": "ai", "content": "Hello Walter White, how are you doing?"},
    {"role": "user", "content": "Say my name..."}
]                                                                               # storing history of user-ai chat

res = llm.invoke(prompts)
print(res.content)  # model understand the context of walter white and breaking bad show
```

### Message History (While Loop)

```python
prompts = []

while True:
    query = input("User: ")

    if query.lower() in ['stop', 'bye', 'exit', 'esc']:
        break;
   
    prompts.append({"role": "user", "content": query})
    res = llm.invoke(prompts)
    prompts.append({"role":"ai", "content": res.content})      # ai (not AI) langchain has roles assigned already
    print(res.content, '\n')
```
