```python
from dotenv import load_dotenv
load_dotenv()

from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model = 'gpt-4o')
```

### Structured Output

```python
text = "My name is Walter White and Email is walterwhite_0289@breakingbad.com and my age is 52. Also, my personality trait is Narcissistic"

# This is text, from which we want the llm to store values of name, email age and trait in seperate variables so we can call them
```

```python
from pydantic import BaseModel, Field
from typing import List

class ResponseStructure(BaseModel):
    name:str = Field(description= "Name of the person")
    age:str = Field(description= "Age of the Person")
    email:str = Field(description="Age of the Person")
    trait:str = Field(description= "Personality trait of the person")

structured_llm = llm.with_structured_output(ResponseStructure)         # creating a new sub-llm that outputs only in the given format
```

```python
res = structured_llm.invoke(text)

res # Here the output of the structured llm is stored in the object tyle we created (ResponseStructure)
```

```python
print(res.name,"\n",res.trait)   #  Different varibales now can be called 
```

```python
res.model_dump()
```

```python
class Movies(BaseModel):
    name:str = Field("Name of the Movie")
    year:str = Field("Year of Release")
    genre:str = Field("Genre of the movie")

class AllMovies(BaseModel):
    movies:List[Movies] = Field("List of all movies")               # List[Movies] gets a list of Movies type obj (our pydantic creation)

movie_llm = llm.with_structured_output(AllMovies)                   # stores list of movies data
```

```python
res = movie_llm.invoke("Give me three random movies")
```

```python
res.model_dump()
```

```python
print(res.movies[0].name, res.movies[1].name) # Accesing from the pydantic list
```
